```python
data = pd.read_json("merged\mergedalldata.json")
newcarjson = data['new_car_overview']
newcarjson
```

```python
df_newcarjson = pd.DataFrame(newcarjson)
df_newcarjson
```

```python
import ast
import json

df_newcarjson['new_car_overview'] = df_newcarjson['new_car_overview'].apply(ast.literal_eval)

# Extract the 'top' data and create columns from the keys
top_data = df_newcarjson['new_car_overview'].apply(lambda x: {item['key']: item['value'] for item in x['top']})

# Convert the dictionary of 'top' data to a DataFrame
merged_second_column = pd.DataFrame(top_data.tolist())

# Display the resulting DataFrame
merged_second_column
```

Taking the third column to analysis

```python
import json
data = pd.read_json("merged\mergedalldata.json")
carfeature = data['new_car_feature']
carfeature
```

```python
df_carfeature = pd.DataFrame(carfeature)
df_carfeature
```

```python
# Ensure 'new_car_feature' is a dictionary (convert from string if needed)
if df_carfeature['new_car_feature'].apply(type).eq(str).any():
    df_carfeature['new_car_feature'] = df_carfeature['new_car_feature'].apply(ast.literal_eval)

# Initialize an empty dictionary to hold column data
columns = {}

# Process the 'top' data
df_carfeature['top'] = df_carfeature['new_car_feature'].apply(lambda x: x.get('top', []))
columns['Features'] = df_carfeature['top'].apply(len)

# Process the 'data' field
df_carfeature['data'] = df_carfeature['new_car_feature'].apply(lambda x: x.get('data', []))
for entry in df_carfeature['data'].values[0]:
    heading = entry['heading']
    list_length = len(entry['list'])
    columns[f'{heading}'] = [list_length] * len(df_carfeature)

# Convert the dictionary to a DataFrame
car_feature = pd.DataFrame(columns)

# Display the DataFrame
car_feature
```

```python
df_newcar_detail = pd.concat([df_new_car_1,df_newcar2,merged_second_column,car_feature], axis =1)
df_newcar_detail
```

Taking the fourth column to analysis

```python
car_specific=data['new_car_specs']
car_specific
```

```python
car_specific1=pd.DataFrame(car_specific)
car_specific1
```

```python
import ast
# Ensure 'new_car_specs' column is a dictionary
car_specific1['new_car_specs'] = car_specific1['new_car_specs'].apply(ast.literal_eval)

# Normalize the 'new_car_specs' data directly
normalized_data = car_specific1['new_car_specs'].apply(
    lambda specs: {
        **{item['key']: item['value'] for item in specs['top']},
        **{item['key']: item['value'] for section in specs['data'] for item in section['list']}
    }
)

# Convert to DataFrame
car_specific = pd.json_normalize(normalized_data)

car_specific
```
