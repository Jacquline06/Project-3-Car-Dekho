```python
def convert_height(height):
    if isinstance(height, float):  # If already a float, just return it
        return height
    
    # Convert to string if it's not already
    height_str = str(height)
    
    # Remove non-numeric characters and extra spaces
    height_str = height_str.replace("mm", "").replace(",", "").strip()
    
    # Check if the string contains a range
    if '-' in height_str:
        # Split the range and calculate the average
        min_value, max_value = height_str.split('-')
        return (float(min_value) + float(max_value)) / 2
    else:
        # Convert directly to float if it's not a range
        return float(height_str)

# Apply the function to the column
df['Height'] = df['Height'].apply(convert_height)

```

```python
import pandas as pd
import re
def extract_rpm(torque):
    if isinstance(torque, str):  # Ensure the value is a string
        # Use regular expression to find the RPM value
        rpm_match = re.search(r'(\d+)rpm', torque)
        if rpm_match:
            return int(rpm_match.group(1))
    return None

# Apply the function to the column
df['RPM'] = df['Max Torque'].apply(extract_rpm)



```

```python
def extract_bhp(power_str):
    if isinstance(power_str, str):  # Ensure the value is a string
        # Use regular expression to find the `bhp` value
        bhp_match = re.search(r'(\d+\.\d+)bhp', power_str)
        if bhp_match:
            return float(bhp_match.group(1))
    return None

# Apply the function to the column
df['BHP'] = df['Max Power'].apply(extract_bhp)

```

```python
def clean_numeric_column(column, unit):
    # Convert to string to ensure .str methods work
    column = column.astype(str)
    # Remove specified unit and any extra characters like quotes or spaces
    column = column.str.replace(unit, '', regex=False)
    column = column.str.replace('"', '', regex=False)  # Remove quotation marks if present
    column = column.str.replace(' ', '', regex=False)  # Remove any extra spaces
    return pd.to_numeric(column, errors='coerce')

# Clean and convert 'Wheel Size' column
df['Wheel Size'] = clean_numeric_column(df['Wheel Size'], ' inches')

# Clean and convert 'BHP' column
df['BHP'] = clean_numeric_column(df['BHP'], ' bhp')

# Calculate the mean of the numeric values and fill missing values
mean_wheel_size = df['Wheel Size'].mean()
df['Wheel Size'] = df['Wheel Size'].fillna(value=mean_wheel_size)

mean_bhp = df['BHP'].mean()
df['BHP'] = df['BHP'].fillna(value=mean_bhp)
```
